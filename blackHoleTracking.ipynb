```python
def trackBH(infile,head=True):
    lines = []
    prevBH = []
    bhList = []
    startline = 0
    inline = infile.readline()
    if head == True:
        inline = infile.readline()
    while inline:
        lines.append(inline)
        inline = infile.readline()
    for i in range(len(lines)):
        # reset variables
        type3 = -1
        bh1 = False
        bh2 = False
        bh3 = False
        id3 = -1
        currBH = [-1,0,0]
        mass1 = 0
        mass2 = 0
        mass3 = 0
        
        # split line
        sline = lines[i].split(" ")
            
        # find bh to track
        ## get type of object
        for elem in sline[-3:]:
            if elem[:5] == 'type1':
                type1 = int(elem[6:])
            if elem[:5] == 'type2':
                type2 = int(elem[6:])
            if elem[:5] == 'type3':
                type3 = int(elem[6:])
        
        ## search lines[i] for a bh
        if type1 == 14:
            bh1 = True
        if type2 == 14:
            bh2 = True
        if type3 == 14:
            bh3 = True
            
        # get bh ID
        ##split component information
        info = sline[3].split(":")
        
        ## get all IDs
        if bh1 == True:
            sinfo = info[0].split("(")
            id1 = int(sinfo[0][4:])
        if bh2 == True:
            sinfo = info[1].split("(")
            id2 = int(sinfo[0][4:])
        if bh3 == True:
            sinfo = info[2].split("(")
            id3 = int(sinfo[0][4:])
        
        # update bh ID
        if bh1 == True and id1 not in prevBH:
            currBH[0] = id1
        elif bh2 == True and id2 not in prevBH:
            currBH[0] = id2
        elif bh3 == True and id3 not in prevBH:
            currBH[0] = id3
        
        # update companion mass encountered
        ## get masses
        if bh1 == False:
            sinfo = info[0].split("(")
            mass1 = float(sinfo[1][3:-1])
        if bh2 == False:
            sinfo = info[1].split("(")
            mass2 = float(sinfo[1][3:-1])
        if bh3 == False and id3 >= 0:
            sinfo = info[2].split("(")
            mass3 = float(sinfo[1][3:-1])
        
        ##add masses
        currBH[1] = currBH[1] + mass1 + mass2 + mass3
        
        # add 1 to collision counter
        currBH[2] += 1
        
        # archive bh used
        prevBH.append(currBH[0])
        
        # update bh id with merger product id
        ## split merger information
        minfo = sline[2].split("(")
        
        ## get merger ID
        idm = int(minfo[0][4:])
        
        ## update currBH ID
        currBH[0] = idm
        
        # search rest of lines for id as an input to a merger
        for j in range(i+1,len(lines)):
            # reset variables
            type3 = -1
            bh1 = False
            bh2 = False
            bh3 = False
            id1 = -1
            id2 = -1
            id3 = -1
            match1 = False
            match2 = False
            match3 = False
            mass1 = 0
            mass2 = 0
            mass3 = 0
            
            # split line
            sline = lines[j].split(" ")

            # find id of tracked bh
            ## get type of object
            for elem in sline[-3:]:
                if elem[:5] == 'type1':
                    type1 = int(elem[6:])
                if elem[:5] == 'type2':
                    type2 = int(elem[6:])
                if elem[:5] == 'type3':
                    type3 = int(elem[6:])
            
            ## check for bh
            if type1 == 14:
                bh1 = True
            if type2 == 14:
                bh2 = True
            if type3 == 14:
                bh3 = True
                
            ## get bh id
            if bh1 == True:
                sinfo = info[0].split("(")
                id1 = int(sinfo[0][4:])
            if bh2 == True:
                sinfo = info[1].split("(")
                id2 = int(sinfo[0][4:])
            if bh3 == True:
                sinfo = info[2].split("(")
                id3 = int(sinfo[0][4:])
                
            ## compare
            if id1 == currBH[0]:
                match1 = True
            elif id2 == currBH[0]:
                match2 = True
            elif id3 == currBH[0]:
                match3 = True
                
            # update companion mass encountered
            ## get masses
            if bh1 == False and match1 == True:
                sinfo = info[0].split("(")
                mass1 = float(sinfo[1][3:-1])
            if bh2 == False and match2 == True:
                sinfo = info[1].split("(")
                mass2 = float(sinfo[1][3:-1])
            if bh3 == False and match3 == True:
                sinfo = info[2].split("(")
                mass3 = float(sinfo[1][3:-1])
        
            ##add masses
            currBH[1] = currBH[1] + mass1 + mass2 + mass3
            
            if match1 == True or match2 == True or match3 == True:
                # add 1 to collision encounter
                currBH[2] += 1
            
                # update bh id with merger product id
                ## split merger information
                minfo = sline[2].split("(")
                
                ## get merger ID
                idm = int(minfo[0][4:])
        
                ## update currBH ID
                currBH[0] = idm            
            
                # archive bh used
                prevBH.append(currBH[0])
        # archive bh used
        prevBH.append(currBH[0])
        
        # add finished bh to bhList
        bhList.append(currBH)
        
    # return bhList
    return bhList
```
